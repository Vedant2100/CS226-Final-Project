```python
region = ee.Geometry.Rectangle([-105.7, 39.8, -104.9, 40.3])
```

```python
dataset = (
    ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
    .filterBounds(region)
    .filter(ee.Filter.lt("CLOUDY_PIXEL_PERCENTAGE",20))
    .select(["B4","B8","B11"])
)
```

```python
import os
os.makedirs("/content/data", exist_ok=True)
```

```python
import datetime
import geemap

start = datetime.date(2015,1,1)
end   = datetime.date(2026,3,1)

step_days = 5

current = start

while current < end:

    next_date = current + datetime.timedelta(days=step_days)

    collection = dataset.filterDate(str(current), str(next_date))

    if collection.size().getInfo() > 0:

        image = collection.median().clip(region)

        filename = f"veg_{current}.tif"
        local_path = f"/content/data/{filename}"

        geemap.ee_export_image(
            image,
            filename=local_path,
            scale=52,
            region=region
        )

        s3.upload_file(
            local_path,
            "vegetation-anomaly-cogs",
            f"new_raw/{filename}"
        )

        print("Uploaded:", filename)

    current = next_date
```
